```python
import pandas as pd
import tensorflow as tf
import pandas as pd
import numpy as np
cars_data=pd.read_csv('./vehicles.csv')
cars_data=cars_data.drop(columns=['id','url','image_url','region_url','county','size','vin','lat','long'])
cars_data=cars_data.dropna()
cars_data=cars_data.reset_index()
cars_data=cars_data.drop(['index'],axis = 1)
```

```python
cars_data['condition'] = cars_data['condition'].astype('category')
```

```python
cars_data.head()
```

```python
cars_data['condition'] = cars_data.condition.cat.codes
```

```python
cars_data.head()
```

```python
target = cars_data.pop('condition')
```

```python
cars_learn_data=cars_data['description']
```

```python
import nltk
from nltk.corpus import movie_reviews
import random
```

```python
dataset = tf.data.Dataset.from_tensor_slices((cars_data['price'], target.values))
```

```python
for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))
```

```python
train_dataset = dataset.shuffle(len(cars_data)).batch(1)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=15)
```
