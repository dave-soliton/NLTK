```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
required_cols=['Country','Population','Coastline (coast/area ratio)','GDP ($ per capita)','Literacy (%)','Birthrate','Region']
```

```python
countries= pd.read_csv('countries_of_the_world.csv', usecols = required_cols, decimal = ',')
```

```python
countries.head()
```

```python
countries.rename(columns = {'GDP ($ per capita)':'PerCapitaGDP',
                            'Coastline (coast/area ratio)':'CoastlineAreaRatio',
                            'Literacy (%)':'LiteracyRate'},
                 inplace=True)
```

```python
countries=countries.dropna()
```

```python
countries.Region.unique()
```

```python
avg_by_region = countries.groupby(by = 'Region').mean()
```

```python
avg_by_region.Birthrate.plot(kind='barh',title='Av')
plt.show()
```

```python
pop_by_region = countries[['Region','Population']].groupby(by='Region').sum()
```

```python
pop_by_region.plot.pie(y='Population',figsize=(8,8),legend=False,cmap='Set3')
plt.show()
```

```python
countries['Birthrate'].plot(kind='hist',bins=20)
plt.show()
```

```python
countries.plot.scatter(x='PerCapitaGDP',y='LiteracyRate')
plt.show()
```
